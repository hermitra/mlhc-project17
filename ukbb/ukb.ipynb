# Analysis of tx_ukb671373_long.txt

```python
from tqdm import tqdm
import numpy as np
import pandas as pd
import pickle
```

```python
def missing_dictionary(df, miss_dic):
    for col in df.columns:
        miss_dic[col] += df[col].isnull().sum()
    return miss_dic
```

```python
# load tx_ukb671373_long.txt
chunksize = 10
with pd.read_csv('tx_ukb671373_long.txt', sep='\t', chunksize=chunksize) as reader:
    for chunk in tqdm(reader):
        miss_dic = {col:0 for col in chunk.columns}
        break
```

```python
# load tx_ukb671373_long.txt
chunksize = 10 ** 4
with pd.read_csv('tx_ukb671373_long.txt', sep='\t', chunksize=chunksize) as reader:
    for chunk in tqdm(reader):
        miss_dic = missing_dictionary(chunk, miss_dic)
```

```python
# save the dictionary
with open('miss_dic.pickle', 'wb') as handle:
    pickle.dump(miss_dic, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
# load the dictionary
with open('miss_dic.pickle', 'rb') as handle:
    miss_dic = pickle.load(handle)
```

```python
no_of_samples = 200590
miss_dic_percent = {k: v / no_of_samples for k, v in miss_dic.items()}
```

```python
chunksize = 10 ** 4
with pd.read_csv('tx_ukb671373_long.txt', sep='\t', chunksize=chunksize) as reader:
    for chunk in reader:
        tx = chunk
        break
```

```python
# remove columns with more than 50% missing values using the dictionary
tx2 = tx.drop([k for k, v in miss_dic_percent.items() if v > 0.01], axis=1)
```

```python
columns_coded = [k for k, v in miss_dic_percent.items() if v <= 0.01]
# remove f. prefix
columns_coded = [col[2:] for col in columns_coded[1:]] # also remove the first column (f.eid)
# only first . is replaced with -
columns_coded = [col.replace('.', '-', 1) for col in columns_coded]
columns_coded
```

```python
# load html file that contains the coding information (column_encoding_ukb671373.html)
# only read first table
coding = pd.read_html('column_encoding_ukb671373.html')
```

```python
coding[0].head()
```

```python
columns_descr = []
for col in columns_coded:
    dscr = coding[0].query('UDI == "{}"'.format(col))['Description'].values[0]
    columns_descr.append(dscr + ' (' + col + ')')
```

```python
columns_descr2 = ['eid'] + columns_descr
columns_descr2
```

```python
# save the list in text file
with open('columns_descr.txt', 'w') as f:
    for item in columns_descr2:
        f.write("%s\n" % item)
```
