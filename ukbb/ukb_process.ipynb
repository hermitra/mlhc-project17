```python
import pandas as pd
import numpy as np
```

```python
pheno = pd.read_csv('tx_ukb48826.txt', sep='\t')
pheno = pheno.drop(pheno.columns[np.arange(1, 7)], axis=1)
pheno
```

```python
pheno = pheno.iloc[:, :-16]
pheno
```

```python
threshold = pheno.shape[1]

# drop columns with more than 50% missing values
df = pheno.dropna(thresh=threshold, axis=0)

# print the resulting dataframe
for col in df.columns:
    if col.count('.') == 3:  # check if the column name matches the pattern
        dot_pos = col.find('.')  # find the position of the first dot
        new_col = col[:dot_pos+1] + col[dot_pos+1:].replace('.', '-', 1)  # replace the second dot with a hyphen
        df = df.rename(columns={col: new_col})
df
```

```python
prefixes = set([col.rsplit('-', maxsplit=1)[0] for col in df.columns[1:]])

# compute the average of the columns with the same prefix and create a new column with the result
for prefix in prefixes:
    cols_to_avg = [col for col in df.columns[1:] if col.startswith(prefix)]
    df[prefix + '-0.0'] = df.loc[:, cols_to_avg].mean(axis=1)
    df = df.drop(cols_to_avg[1:], axis=1)

# print the resulting dataframe
df
```

```python
coding = pd.read_html('MLHC17/UKBB htmls/ukb48826.html')
```

```python
columns_descr = []
for col in [i[2:] for i in df.columns.tolist()[1:]]:
    dscr = coding[1].query('UDI == "{}"'.format(col))['Description'].values[0]
    columns_descr.append(dscr + ' (' + col + ')')
```

```python
df.columns = ['IID'] + columns_descr
```

```python
df.insert(0, 'FID', df['IID'])
df
```

```python
df.columns =df.columns.str.replace(r'\s+', '_', regex=True)
```
