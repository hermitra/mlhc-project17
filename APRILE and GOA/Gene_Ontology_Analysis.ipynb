```python
import pandas as pd
import numpy as np
from goatools.obo_parser import GODag
from goatools.associations import read_ncbi_gene2go

# Load the GO OBO file
go_obo_file = 'go.obo'
obodag = GODag(go_obo_file)

# Load the NCBI gene2go file
ncbi_gene2go_file = 'gene2go'
gene2go = read_ncbi_gene2go(ncbi_gene2go_file)
```

```python
# Create a dataframe of NCBI Gene IDs
ncbi_ids = np.array(mapped_df[' ncbi_gene_id'].tolist()).astype(int)
df_ncbi = pd.DataFrame({'NCBI': ncbi_ids})

# Create an empty dataframe to store the results
df_result = pd.DataFrame(columns=['NCBI', 'GO'])
go_ids = []
ncbi_ids_list = []
# Loop through each NCBI Gene ID in the dataframe and find its corresponding GO terms
for ncbi_id in ncbi_ids:
    if ncbi_id in gene2go:
        for go_id in gene2go[ncbi_id]:
          go_ids.append(go_id)
          ncbi_ids_list.append(ncbi_id)
          

df_result['NCBI'] =  ncbi_ids_list
df_result['GO'] =  go_ids
df_result = df_result.drop_duplicates(subset=['NCBI'])
# Print the results
print(df_result)
```

```python
from goatools.base import get_godag
godag = get_godag('go-basic.obo', optional_attrs='relationship')
```

```python
goid = 'GO:0007158'
```

```python
def prt_flds(gosubdag):
    """Print the available printing fields"""
    print('Print fields:')
    for fld in sorted(gosubdag.prt_attr['flds']):
        print('    {F}'.format(F=fld))
```

```python
from goatools.gosubdag.gosubdag import GoSubDag

# Create a subset of the GO DAG which contains:
#   * The selected GO term and
#   * All the GO terms above it
gosubdag = GoSubDag(goid, godag, relationships=True, prt=False)

# Get additional information for chosen GO
ntgo = gosubdag.go2nt[goid]

# Choose fields and custom printing format
# prt_flds(gosubdag)  # Uncomment to see the available print fields
prtfmt = '{NS} {GO} D{depth:02} {GO_name}'

# Print detailed information for GO
print(prtfmt.format(**ntgo._asdict()))
```
